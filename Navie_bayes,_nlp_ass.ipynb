```python
import re  # Import the 're' module for regular expressions

def clean_text(text):
  text = text.lower()  # Lowercase
  text = re.sub(r"[^\w\s]", "", text)  # Remove punctuation
  # Add more steps like removing stopwords
  return text

blog["Data"] = blog["Data"].apply(clean_text)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer  # Import TfidfVectorizer

vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(blog["Data"])
```

```python
features
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, blog["Labels"], test_size=0.2)
```

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sentiment_analyzer = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
  scores = sentiment_analyzer.polarity_scores(text)
  sentiment = "neutral"
  if scores['pos'] > 0.5:
    sentiment = "positive"
  elif scores['neg'] > 0.5:
    sentiment = "negative"
  return sentiment

blog["Sentiment"] = blog["Data"].apply(analyze_sentiment)
```

```python
blog['Sentiment']
```

```python
sentiment_by_category = blog.groupby("Labels")["Sentiment"].value_counts().unstack(fill_value=0)
print(sentiment_by_category)
```

```python
from textblob import TextBlob

def analyze_sentiment_textblob(text):
  sentiment = TextBlob(text).sentiment
  if sentiment.polarity > 0.5:
    sentiment = "positive"
  elif sentiment.polarity < -0.5:
    sentiment = "negative"
  else:
    sentiment = "neutral"
  return sentiment
```

```python
blog["Sentiment_Vader"] = blog["Data"].apply(analyze_sentiment_textblob)

# Sentiment Distribution by Category (using Vader scores)
sentiment_by_category_vader = blog.groupby("Labels")["Sentiment_Vader"].value_counts().unstack(fill_value=0)
print(sentiment_by_category_vader)
```
